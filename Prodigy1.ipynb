```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
file_path = 'world_population.csv'  # Update this to your file path if necessary
df = pd.read_csv(file_path)

# Set the style for the plots
sns.set(style="whitegrid")

# 1. Histogram of 2022 Population Distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['2022 Population'], bins=30, color='blue')
plt.title('Distribution of 2022 Population Across Countries')
plt.xlabel('Population')
plt.ylabel('Number of Countries')
plt.show()

# 2. Bar Chart of the Top 10 Most Populous Countries in 2022
top_10_countries = df.nlargest(10, '2022 Population')[['Country/Territory', '2022 Population']]

plt.figure(figsize=(12, 6))
sns.barplot(x='2022 Population', y='Country/Territory', data=top_10_countries, palette='viridis')
plt.title('Top 10 Most Populous Countries in 2022')
plt.xlabel('Population')
plt.ylabel('Country/Territory')
plt.show()
```
